Modify the previous program to use 5 quadrature points instead of two.

Integrate the functions with the 2-point and the 5-point quadrature rules.

Compare the results with the analytical values.

```python
import numpy as np
import time
```

```python
#translation of the original FORTRAN code
nq=2
xq=(-1/np.sqrt(3), 1/np.sqrt(3))
wq=(1,1)
I=0

def fct(x):
  return x**2

for iq in range(nq):
  I=I+wq[iq]*fct(xq[iq])

print('I=', I)
```

```python
#modification to the 5 quadrature points
nq=5
xq=(0, np.sqrt(5-2*np.sqrt(10/7))/3, -np.sqrt(5-2*np.sqrt(10/7))/3, np.sqrt(5+2*np.sqrt(10/7))/3, -np.sqrt(5+2*np.sqrt(10/7))/3)
wq=(128/225, (322 + 13*np.sqrt(70))/900, (322 + 13*np.sqrt(70))/900, (322 - 13*np.sqrt(70))/900, (322 - 13*np.sqrt(70))/900)
I=0

def fct(x):
  return x**2

for iq in range(nq):
  I=I+wq[iq]*fct(xq[iq])

print('I=', I)
```

```python
print(f'The difference in the results via using 5 quadrature points Gaussian quadrature algorithm instead of two is negligable')
```

```python
def the_quadrature_rules_2(func):
  nq=2
  xq=(-1/np.sqrt(3), 1/np.sqrt(3))
  wq=(1,1)
  I=0

  for iq in range(nq):
    I=I+wq[iq]*func(xq[iq])
    
  return I
```

```python
def the_quadrature_rules_5(func):
  nq=5
  xq=(0, np.sqrt(5-2*np.sqrt(10/7))/3, -np.sqrt(5-2*np.sqrt(10/7))/3, np.sqrt(5+2*np.sqrt(10/7))/3, -np.sqrt(5+2*np.sqrt(10/7))/3)
  wq=(128/225, (322 + 13*np.sqrt(70))/900, (322 + 13*np.sqrt(70))/900, (322 - 13*np.sqrt(70))/900, (322 - 13*np.sqrt(70))/900)
  I=0

  for iq in range(nq):
    I=I+wq[iq]*func(xq[iq])

  return I
```

```python
time1= time.time()
qr2_sin=the_quadrature_rules_2(lambda x: np.sin(np.pi*x + np.pi/2))
time2= time.time()
qr2_sqrt=the_quadrature_rules_2(lambda x: np.sqrt(x+1))
time3= time.time()
qr2_cub=the_quadrature_rules_2(lambda x: x**4 - x**3)
time4= time.time()

qr5_sin=the_quadrature_rules_5(lambda x: np.sin(np.pi*x + np.pi/2))
time5= time.time()
qr5_sqrt=the_quadrature_rules_5(lambda x: np.sqrt(x+1))
time6= time.time()
qr5_cub=the_quadrature_rules_5(lambda x: x**4 - x**3)
time7= time.time()

start=-1
end=1

anal_sin= np.sin(np.pi*end)/np.pi - np.sin(np.pi*start)/np.pi
anal_sqrt=2*((end+1)**1.5)/3 - 2*((start+1)**1.5)/3
anal_cub=(end**5/5 - end**4/4) - (start**5/5 - start**4/4)
```

```python
print(round(anal_sin,18), round(anal_sqrt,4), round(anal_cub,4))
print(round(qr5_sin,6), round(qr5_sqrt,4), round(qr5_cub,4))
print(round(qr2_sin,4), round(qr2_sqrt,4), round(qr2_cub,4))
```

```python
print(f'''For the analysis of the Gaussian quadrature rule we use three functions:
for the func=sin(pi*x + pi/2) the error for 2 point is {round(abs(anal_sin - qr2_sin),5)}   and for 5 point is {round(abs(anal_sin - qr5_sin),6)};
for the func=sqrt(x+1)        the error for 2 point is {round(abs(anal_sqrt - qr2_sqrt),5)}   and for 5 point is {round(abs(anal_sqrt - qr5_sqrt),5)};
for the func=x**4 - x**3      the error for 2 point is {round(abs(anal_cub - qr2_cub),5)}   and for 5 point is {round(abs(anal_cub - qr5_cub),18)}.''')
```
