Compute analytically the following integral:
f(x,y)= x**2 +4y over the domain Ω = [11, 14] × [7, 10]

Write a code which integrates this function by means of a 2x2, 3x3 or 4x4 Gauss-Legendre quadrature algorithm.

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
import math
import time

sy.init_printing()
```

```python
func=lambda x,y: x**2 + 4*y

a,b,c,d=11,14,7,10
```

```python
x, y = sy.symbols('x y', real=True)

anal=float(sy.integrate(sy.integrate(x**2 + 4*y, (x, a, b)), (y, c, d)))
#(d-c)*b**3/3 + (2*d**2 - 2*c**2)*b - (d-c)*a**3/3 - (2*d**2 - 2*c**2)*a
```

```python
time1=time.time()
nq2=2
r2, s2=(-1/np.sqrt(3), 1/np.sqrt(3)), (-1/np.sqrt(3), 1/np.sqrt(3))
wq2=(1,1)
I2=0

for iqx in range(nq2):
  for iqy in range(nq2):
    xq2= a + (1+r2[iqx])*(b-a)/2
    yq2= c + (1+s2[iqy])*(d-c)/2
    I2=I2 + 0.5*(d-c)*0.5*(b-a)*wq2[iqy]*func(xq2,yq2)*wq2[iqx]
time2=time.time()
```

```python
time3=time.time()
nq3=3
r3, s3=(0, np.sqrt(3/5), -np.sqrt(3/5)), (0, np.sqrt(3/5), -np.sqrt(3/5))
wq3=(8/9, 5/9, 5/9)
I3=0

for iqx in range(nq3):
  for iqy in range(nq3):
    xq3= a + (1+r3[iqx])*(b-a)/2
    yq3= c + (1+s3[iqy])*(d-c)/2
    I3=I3 + 0.5*(d-c)*0.5*(b-a)*wq3[iqy]*func(xq3,yq3)*wq3[iqx]
time4=time.time()
```

```python
time5=time.time()
nq4=4
r4=(np.sqrt(3/7 - 2*np.sqrt(6/5)/7), -np.sqrt(3/7 - 2*np.sqrt(6/5)/7), np.sqrt(3/7 + 2*np.sqrt(6/5)/7), -np.sqrt(3/7 + 2*np.sqrt(6/5)/7))
s4=r4
wq4=((18+np.sqrt(30))/36, (18+np.sqrt(30))/36, (18-np.sqrt(30))/36, (18-np.sqrt(30))/36)
I4=0

for iqx in range(nq4):
  for iqy in range(nq4):
    xq4= a + (1+r4[iqx])*(b-a)/2
    yq4= c + (1+s4[iqy])*(d-c)/2
    I4=I4 + 0.5*(d-c)*0.5*(b-a)*wq4[iqy]*func(xq4,yq4)*wq4[iqx]
time6=time.time()
```

```python
time7=time.time()
nq5=5
r5=(0, np.sqrt(5-2*np.sqrt(10/7))/3, -np.sqrt(5-2*np.sqrt(10/7))/3, np.sqrt(5+2*np.sqrt(10/7))/3, -np.sqrt(5+2*np.sqrt(10/7))/3)
s5=r5
wq5=(128/225, (322 + 13*np.sqrt(70))/900, (322 + 13*np.sqrt(70))/900, (322 - 13*np.sqrt(70))/900, (322 - 13*np.sqrt(70))/900)
I5=0

for iqx in range(nq5):
  for iqy in range(nq5):
    xq5= a + (1+r5[iqx])*(b-a)/2
    yq5= c + (1+s5[iqy])*(d-c)/2
    I5=I5 + 0.5*(d-c)*0.5*(b-a)*wq5[iqy]*func(xq5,yq5)*wq5[iqx]
time8=time.time()
```

```python
error=[(I2-anal), (I3-anal), (I4-anal), (I5-anal)]
print(f'''The Gauss-Legendre quadrature algorithm by 
2x2 quadrature points results in {round(error[0],17)} difference from the analytical solution, 
3x3 quadrature points results in {round(error[1],17)} difference from the analytical solution, 
4x4 quadrature points results in {round(error[2],17)} difference from the analytical solution, and 
5x5 quadrature points results in {round(error[3],17)} difference from the analytical solution.''')
```

```python
print('The solutions for 2x2, 3x3, 4x4, and 5x5 Gauss-Legendre quadrature algorithm are as follows\n', *[I2, I3, I4, I5])
      #*list(map(lambda x: round(x,8), [I2, I3, I4, I5])))
```
