Compute the analytical values of M, Ka and Kd for rectangular elements of size hx , hy .

```python
import numpy as np
# import matplotlib as mpl
# import matplotlib.pyplot as plt
# import time
# import scipy
# import scipy.linalg
# import random
# import pandas as pd
# import seaborn as sns
import sympy as sy
import math

sy.init_printing()
```

# matrixes

```python
hx, hy, vx, vy, r, s = sy.symbols('hx hy vx vy r s', real=True)
```

```python
B= sy.Matrix()

N=sy.Matrix([sy.sympify(('(1+', '(1-')[not k%3] + 'r)' + '*'  + ('(1+', '(1-')[k<2] + 's)' + str('/4')) for k in range(4)]).T

for j in range(4):
  step1=[]
  for i in range(2):
    first=('s)/hx/2', 'r)/hy/2')[i]
    second=['(-1+', '(1-', '(1+', '(-1-']
    step1.append(second[j*((-1)**i)] + first)
  B= B.col_insert(j, sy.Matrix(step1))
```

```python
BT=B.T
NT=N.T
```

```python
N
```

```python
wow = [str((1 + r*(-1)**math.ceil(np.cos(k*np.pi/2 +np.pi/4))) * (1 
      + s*(-1)**math.ceil(np.cos(k*np.pi/2 -np.pi/4))) / 4)  for k in range(4)]
sy.Matrix(wow).T
```

```python
B_V=sy.Matrix([vx*B[k] + vy*B[k+4] for k in range(4)]).T
```

```python
Ka_matrix=NT*B_V
Ka_matrix
```

```python
Ka_matrix_integrated = Ka_matrix[:, :]

for k in range(16):
  result=sy.integrate(sy.integrate(Ka_matrix[k], ('r', -1, 1)), ('s', -1, 1))
  Ka_matrix_integrated[k]=result.expand()

Ka_matrix_integrated
```

```python
Kd_matrix=BT*B
Kd_matrix
```

```python
Kd_matrix_integrated = Kd_matrix[:, :]

for k in range(16):
  result=sy.integrate(sy.integrate(Kd_matrix[k], ('r', -1, 1)), ('s', -1, 1))
  Kd_matrix_integrated[k]=result.expand()

Kd_matrix_integrated
```

```python
M_matrix=NT*N
M_matrix
```

```python
M_matrix_integrated = M_matrix[:, :]

for k in range(16):
  result=sy.integrate(sy.integrate(M_matrix[k], ('r', -1, 1)), ('s', -1, 1))
  M_matrix_integrated[k]=result.expand()

M_matrix_integrated
```

# solution

```python
Cp=23
rho=1000
hx1=3
hy1=2
vx1=4
vy1=7
k=5
```

```python
M=rho*Cp*hx1*hy1* M_matrix_integrated /4
M.evalf()
```

```python
nq=2
r, s=(-1/np.sqrt(3), 1/np.sqrt(3)), (-1/np.sqrt(3), 1/np.sqrt(3))
wq=(1,1)

I_M, I_Ka, I_Kd = 0,0,0
v_vector=np.array([vx1,vy1])

for i in range(nq):
  for j in range(nq):
    N=np.array([(1-r[i])*(1-s[j])/4, (1+r[i])*(1-s[j])/4, 
                (1+r[i])*(1+s[j])/4, (1-r[i])*(1+s[j])/4])
    NT=N.reshape((4,1))
    B=np.array([[(s[j]-1)/2/hx1, (1-s[j])/2/hx1, (1+s[j])/2/hx1, (-1-s[j])/2/hx1], 
               [(r[i]-1)/2/hy1, (-r[i]-1)/2/hy1, (r[i]+1)/2/hy1, (1-r[i])/2/hy1]])
    BT=B.T

    I_M+=(np.multiply(NT, N))*wq[i]*wq[j]
    I_Ka+=(np.multiply(NT, np.dot(v_vector, B)))*wq[i]*wq[j]
    I_Kd+=(np.dot(BT, B))*wq[i]*wq[j]
  
I_M*=rho*Cp*hx1*hy1/4
I_Kd*=k*hx1*hy1/4
I_Ka*=rho*Cp*hx1*hy1/4

I_Kd
```

```python
Kd_matrix_solved=Kd_matrix_integrated.subs('hy', hy1).subs('hx', hx1)

Kd= k*hx1*hy1* Kd_matrix_solved/4

Kd.evalf()
```

```python
Ka_matrix_solved=Ka_matrix_integrated.subs('hy', hy1).subs('hx', hx1).subs(vx, vx1).subs(vy, vy1)   #why on Earth there is no '' for velocity....

Ka= rho*Cp*hx1*hy1* Ka_matrix_solved/4

Ka.evalf()
```
